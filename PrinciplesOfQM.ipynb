```python
%%manim -p -ql -v WARNING Principles

from manim import *

class Principles(Scene):
    def construct(self):
        # 1. Short equation
        eq_short = MathTex(r"\hat{H}\Psi = E\Psi", color=TEAL).scale(1).shift(UP*1.5)
        eq_two = MathTex(r"(\hat{T}+\hat{V})\Psi = E\Psi", color=TEAL).scale(1).shift(UP*1.5)
        eq_three = MathTex(r"(\hat{T}+V(X))\Psi = E\Psi", color=TEAL).scale(1).shift(UP*1.5)

        eq_T_one = MathTex(r"\hat{T} = \frac{\hat{p}^2}{2m}", color=YELLOW).scale(1).shift(UP*0)
        eq_T_two = MathTex(r"\hat{p} = -i\hbar\frac{d}{dx}", color=ORANGE).scale(1).shift(DOWN*1.5)
        eq_T_three = MathTex(r"\hat{T} = \frac{1}{2m}\left(-i\hbar\frac{d}{dx}\right)^2", color=YELLOW).scale(1).shift(UP*0)
        eq_T_four = MathTex(r"\hat{T} = -\frac{\hbar^2}{2m}\frac{d\psi}{dx^2}^2", color=YELLOW).scale(1).shift(UP*0)
        # 2. Full Schr√∂dinger equation
        eq_full = MathTex(
            r"-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2}+V(X)\psi(X)=E\psi(X)",
            color=TEAL,
            substrings_to_isolate=[r"V(X)"]
        ).scale(1).shift(UP*1.5)

        self.play(Write(eq_short))
        self.wait(1)

        self.play(TransformMatchingTex(eq_short, eq_two, transform_mismatches=True))
        self.wait(1)
        self.play(TransformMatchingTex(eq_two, eq_three, transform_mismatches=True))
        self.wait(1)
        self.play(eq_three[0][1:3].animate.set_color(YELLOW))
        self.wait(1)
        self.play(Write(eq_T_one))
        self.wait(1)
        self.play(eq_T_one[0][3:5].animate.set_color(ORANGE))
        self.wait(1)
        self.play(Write(eq_T_two))
        self.wait(1)
        self.play(TransformMatchingTex(eq_T_one, eq_T_three, transform_mismatches=True))
        self.play(FadeOut(eq_T_two))
        self.wait(1)
        self.play(TransformMatchingTex(eq_T_three, eq_T_four, transform_mismatches=True))
        self.wait(1)
        self.play(TransformMatchingTex(eq_three, eq_full, transform_mismatches=True))
        self.play(FadeOut(eq_T_four))
        self.wait(1)

```
