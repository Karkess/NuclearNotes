```python
%%manim -p -ql -v WARNING Principles

from manim import *

class Principles(Scene):
    def construct(self):
        # 1. Short equation
        S_eq_short = MathTex(r"\hat{H}\Psi = E\Psi", color=TEAL).scale(1).shift(UP*1.5)
        S_eq_two = MathTex(r"(\hat{T}+\hat{V})\Psi = E\Psi", color=TEAL).scale(1).shift(UP*1.5)
        S_eq_three = MathTex(r"(\hat{T}+V(x))\Psi = E\Psi", color=TEAL).scale(1).shift(UP*1.5)

        S_eq_T_one = MathTex(r"\hat{T} = \frac{\hat{p}^2}{2m}", color=YELLOW).scale(1).shift(UP*0)
        S_eq_T_two = MathTex(r"\hat{p} = -i\hbar\frac{d}{dx}", color=ORANGE).scale(1).shift(DOWN*1.5)
        S_eq_T_three = MathTex(r"\hat{T} = \frac{1}{2m}\left(-i\hbar\frac{d}{dx}\right)^2", color=YELLOW).scale(1).shift(UP*0)
        S_eq_T_four = MathTex(r"\hat{T} = -\frac{\hbar^2}{2m}\frac{d\psi}{dx^2}^2", color=YELLOW).scale(1).shift(UP*0)
        # 2. Full Schr√∂dinger equation
        S_eq_full = MathTex(
            r"-\frac{\hbar^2}{2m}\frac{d^2\psi}{dx^2}+V(x)\psi(x)=E\psi(x)",
            color=TEAL,
            substrings_to_isolate=[r"V(x)"]
        ).scale(1).shift(UP*1.5)

        self.play(Write(S_eq_short))
        self.wait(1)

        self.play(TransformMatchingTex(S_eq_short, S_eq_two, transform_mismatches=True))
        self.wait(1)
        self.play(TransformMatchingTex(S_eq_two, S_eq_three, transform_mismatches=True))
        self.wait(1)
        self.play(S_eq_three[0][1:3].animate.set_color(YELLOW))
        self.wait(1)
        self.play(Write(S_eq_T_one))
        self.wait(1)
        self.play(S_eq_T_one[0][3:5].animate.set_color(ORANGE))
        self.wait(1)
        self.play(Write(S_eq_T_two))
        self.wait(1)
        self.play(TransformMatchingTex(S_eq_T_one, S_eq_T_three, transform_mismatches=True))
        self.play(FadeOut(S_eq_T_two))
        self.wait(1)
        self.play(TransformMatchingTex(S_eq_T_three, S_eq_T_four, transform_mismatches=True))
        self.wait(1)
        self.play(TransformMatchingTex(S_eq_three, S_eq_full, transform_mismatches=True))
        self.play(FadeOut(S_eq_T_four))
        self.wait(1)

        S_sol_one = MathTex(r"\Psi(x,t)=\psi(x)e^{-i\omega t}",color=TEAL).scale(1).shift(DOWN*0.75)
        S_sol_omega = MathTex(r"\omega = E/\hbar}",color=TEAL).scale(1).shift(DOWN*1.5)
        self.play(Write(S_sol_one))
        self.wait(1)
        self.play(Write(S_sol_omega))
        self.wait(1)
```
