##1.2 Some Introductory **Terminology**

Z - atomic number  
e - magntude of electrons  
Ze - net nuclear charge  
A - Mass number of the species  
N - Neutron Number    
X- Chemical Species  
$^A_Z\mathrm{X}_N$ - Standard nuclide notation

Nuclide - A specific nuclear species  
Nucleon - An individual proton or neutron  
Isotope - Nuclides with the same proton number (Z) but a different neutron number  
Radioisotope - An isotope which is composed of a relatively unstable nucleus  
Isotone - Same number of neutrons (N) but a different Z  
Isobars - Same mass number (A) but different N and Z

## 1.4 Units and Dimensions

Nuclei are on the scale of $~1\times10^{-15}m-7\times10^{-15}m$, or $1-7 \text{ femtometers (fm)}$, sometimes called one fermi, in honor of Enrio Fermi.

Nuclear decay rates range from $10\times10^{-20}$s to thousands or millions of years.

Nuclear energies are typically measured in mega electron volts (or millions of elevtron volts) (MeV), where 1 ev $=1.602\times10^{-19}$J

Masses are given in units of unified atomic mass $u$, where $u$ is defined to be an atom of $^{12}\mathrm{C}$, that is, the mass of $^{12}\mathrm{C}=12u$

The mass energy conversion is $1 u = 931.502$ MeV, given by the famous equation $E=mc^2$
